```python
from bs4 import BeautifulSoup
import pandas as pd
import unicodedata
import itertools
import requests
import time
import re
```

```python
### GATHER URLS
r = requests.get("https://www.churchofjesuschrist.org/study/general-conference/2021/10?lang=eng")
soup = BeautifulSoup(r.content, "html5lib")
print(soup.prettify())
```

```python
links = soup.find_all("a", {"class": "omeqik-0 cxIGgI list-tile listTile-3yP_N"})
urls = ["https://www.churchofjesuschrist.org" + link.get('href') for link in links if re.search("^/study/general-conference/\d{4}/(04|10)/\d{1}", link["href"])]
for url in urls:
    print(url)
```

```python
## SCRAP SPECIFIC TALKS
r = requests.get("https://www.churchofjesuschrist.org/study/general-conference/2021/10/11nelson")
talk = BeautifulSoup(r.content, "html5lib")
print(talk.prettify())
```

```python
title = talk.find('h1').text
title
```

```python
speaker = talk.find('p', {"class": "author-name"}).text
speaker
```

```python
calling = talk.find('p', {"class": "author-role"}).text
calling
```

```python
content_array = talk.find("div", {"class": "body-block"}).find_all("p")
content = ""
for paragraph in content_array:
    content = content + paragraph.text + "\n\n "
content
```
